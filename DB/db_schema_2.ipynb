```python
import psycopg2

hostname = "localhost"
database = "Language_Portal"
username = "postgres"
pwd = ""
port_id = 5432

conn = None
cur = None

try:
    conn = psycopg2.connect(
        host = hostname,
        dbname = database,
        user = username,
        password = pwd,
        port = port_id
    )
    
    cur = conn.cursor()

    insert_script = "INSERT INTO COURSE (LANG, NUM_STUDENTS, NUM_MODULES, NUM_LESSONS) VALUES(%s,%s,%s,%s)"
    insert_values = []
    f = open("courses.txt","r")
    s = f.readline()
    while(s):
        lst = []
        if s[len(s)-1]=='\n':
            s = s[:len(s)-1]
        record = s.split(",")
        for i in record:
            if i.isdigit():
                lst.append(int(i))
            else:
                lst.append(i)
        #print(lst)    
        insert_values.append(tuple(lst))
        s = f.readline()
    f.close()
    for record in insert_values:
        cur.execute(insert_script,record)
    conn.commit()
    

    insert_script = "INSERT INTO MODULES (MODULE_NAME, PARENT_COURSE, NUM_LESSONS, DIFFICULTY) VALUES(%s,%s,%s,%s)"
    insert_values = []
    f = open("modules.txt","r")
    s = f.readline()
    while(s):
        lst = []
        if s[len(s)-1]=='\n':
            s = s[:len(s)-1]
        record = s.split(",")
        for i in record:
            if i.isdigit():
                lst.append(int(i))
            else:
                lst.append(i)
        #print(lst)    
        insert_values.append(tuple(lst))
        s = f.readline()
    f.close()
    for record in insert_values:
        cur.execute(insert_script,record)
    conn.commit()

    insert_script = "INSERT INTO LESSON (LESSON_INDEX, PARENT_MODULE) VALUES(%s,%s)"
    insert_values = []
    f = open("lessons.txt","r")
    s = f.readline()
    while(s):
        lst = []
        if s[len(s)-1]=='\n':
            s = s[:len(s)-1]
        record = s.split(",")
        for i in record:
            if i.isdigit():
                lst.append(int(i))
            else:
                lst.append(i)
        #print(lst)    
        insert_values.append(tuple(lst))
        s = f.readline()
    f.close()
    for record in insert_values:
        cur.execute(insert_script,record)
    conn.commit()

    insert_script = "INSERT INTO QUESTION (QUESTION_INDEX, PARENT_LESSON, QUESTION_TEXT, WRONG_ANSWER_1, WRONG_ANSWER_2, WRONG_ANSWER_3, ANSWER, INFO) VALUES(%s,%s,%s,%s,%s,%s,%s,%s)"
    insert_values = []
    f = open("questions_3.txt","r")
    s = f.readline()
    while(s):
        lst = []
        if s[len(s)-1]=='\n':
            s = s[:len(s)-1]
        record = s.split(",")
        for i in record:
            if i.isdigit():
                lst.append(int(i))
            else:
                lst.append(i)
        print(lst, len(lst))    
        insert_values.append(tuple(lst))
        s = f.readline()
    f.close()
    for record in insert_values:
        cur.execute(insert_script,record)

    conn.commit()

except Exception as error:
    print(error)
finally:
    if cur is not None: 
        cur.close()
    if conn is not None:
        conn.close()

```
