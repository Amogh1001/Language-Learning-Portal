```python
import psycopg2

hostname = "localhost"
database = "Language_Portal"
username = "postgres"
pwd = ""
port_id = 5432

conn = None
cur = None

try:
    conn = psycopg2.connect(
        host = hostname,
        dbname = database,
        user = username,
        password = pwd,
        port = port_id
    )
    
    cur = conn.cursor()

    cur.execute("DROP TABLE IF EXISTS PROGRESS cascade")
    create_script = '''CREATE TABLE IF NOT EXISTS PROGRESS(
    PROGRESS_ID INT PRIMARY KEY
    )'''
    cur.execute(create_script)
    conn.commit()
    
    cur.execute("DROP TABLE IF EXISTS STUDENT cascade")
    create_script = '''CREATE TABLE IF NOT EXISTS STUDENT(
    STUDENT_ID SERIAL PRIMARY KEY,
    USERNAME VARCHAR(100),
    PWD VARCHAR(50)
    )'''
    cur.execute(create_script)
    conn.commit()

    cur.execute("DROP TABLE IF EXISTS COURSE cascade")
    create_script = '''CREATE TABLE IF NOT EXISTS COURSE(
    COURSE_ID SERIAL PRIMARY KEY,
    LANG VARCHAR(100),
    NUM_STUDENTS INT,
    NUM_MODULES INT,
    NUM_LESSONS INT
    )'''
    cur.execute(create_script)
    conn.commit()

    cur.execute("DROP TABLE IF EXISTS MODULES cascade")
    create_script = '''CREATE TABLE IF NOT EXISTS MODULES(
    MODULE_ID SERIAL PRIMARY KEY,
    MODULE_NAME VARCHAR(100),
    PARENT_COURSE INT,
    NUM_LESSONS INT,
    DIFFICULTY VARCHAR(100)
    )'''
    cur.execute(create_script)
    conn.commit()

    cur.execute("DROP TABLE IF EXISTS LESSON cascade")
    create_script = '''CREATE TABLE IF NOT EXISTS LESSON(
    LESSON_ID SERIAL PRIMARY KEY,
    LESSON_INDEX INT,
    PARENT_MODULE INT,
    DIFFICULTY VARCHAR(100)
    )'''
    cur.execute(create_script)
    conn.commit()

    cur.execute("DROP TABLE IF EXISTS QUESTION cascade")
    create_script = '''CREATE TABLE IF NOT EXISTS QUESTION(
    QUESTION_ID SERIAL PRIMARY KEY,
    QUESTION_INDEX INT,
    PARENT_LESSON INT,
    QUESTION_TEXT VARCHAR(255),
    WRONG_ANSWER_1 VARCHAR(255),
    WRONG_ANSWER_2 VARCHAR(255),
    WRONG_ANSWER_3 VARCHAR(255),
    ANSWER VARCHAR(255),
    INFO VARCHAR(255)
    )'''
    cur.execute(create_script)
    conn.commit()

except Exception as error:
    print(error)
finally:
    if cur is not None: 
        cur.close()
    if conn is not None:
        conn.close()

```
